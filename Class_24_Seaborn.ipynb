```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pokemon_data = pd.read_csv('datasets/Pokemon.csv')
pokemon_data.head()
```

```python
# lmplot(), pairplot(), jointplot()
```

```python
sns.lmplot(x = 'Attack Point',
           y = 'Total Point',
           data = pokemon_data)
```

```python
sns.lmplot(x = 'Attack Point',
           y = 'Total Point',
           data = pokemon_data, 
           col = 'Type 1')
```

```python
pokemon_data['Type 1'].value_counts()
```

```python
sns.lmplot(x = 'Attack Point',
           y = 'Total Point',
           data = pokemon_data, 
           hue = 'Type 1')
```

```python
sns.jointplot(x = 'Attack Point',
              y = 'Total Point',
              data = pokemon_data, 
              kind = 'reg')
```

```python
sns.pairplot(pokemon_data, 
             size = 5,
             x_vars = ['Attack Point', 'Speed Point', 'Health Point'],
             y_vars = ['Total Point'], 
             kind = 'reg')
```

# Categorical Plot

```python
# swarmplot, boxplot, violinplot, barplot, factorplot, countplot

tips = sns.load_dataset('tips')
tips.head()

```

```python
# Swarmplot - to get each observation separately

plt.subplots(figsize = (15, 7))
sns.swarmplot(x = 'day',
              y = 'tip',
              data = tips, 
              hue = 'sex')
```

```python
# boxplot - abstract repreentation of data
plt.subplots(figsize = (15, 7))
sns.boxplot(x = 'day',
            y = 'tip',
            data = tips, )
```

```python
plt.subplots(figsize = (15, 7))
sns.violinplot(x = 'day',
            y = 'tip',
            data = tips,)
```

```python
plt.subplots(figsize = (15,5))
sns.violinplot(x ='day',
               y = 'tip',
               data = tips,
               inner = None)
sns.swarmplot(x ='day',
              y ='tip',
              data = tips,
              color ='w')
plt.show()
```

```python
plt.subplots(figsize = (15,5))
sns.barplot(x ='day',
            y = 'tip',
            data = tips,
            hue = 'size')
```

```python
sns.countplot(x = 'day', 
             data = tips,
             hue = 'sex')
```

```python

sns.factorplot(x ='day',
               y = 'tip',
               data = tips,
               kind = 'box',
               size = 5, 
              col = 'sex')
```

```python

sns.factorplot(x ='day',
               y = 'tip',
               data = tips,
               kind = 'violin',
               size = 5, 
              col = 'sex')
```

```python

sns.factorplot(x ='day',
               y = 'tip',
               data = tips,
               kind = 'swarm',
               size = 5, 
              col = 'sex')
```

```python
tips['sex'].value_counts().plot.pie(autopct = '%.2f%%')
```
