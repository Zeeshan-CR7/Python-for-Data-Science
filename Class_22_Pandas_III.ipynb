# Groupby and pivot

```python
car_data = pd.read_csv('Data_sets/cars_data.csv')
car_data
```

```python
list(car_data.groupby('origin_city'))
```

```python
np.min(car_data.groupby('origin_city').get_group('US'))
```

```python
car_data['new_col'] = list(map(lambda val:'low' if val<100 else 'High', car_data['horsepower']))
car_data.groupby('new_col').get_group('low')
```

```python
list(map(lambda val:'low' if val<100 else 'High', car_data['horsepower']))
```
