# Concatenate and merging the dataframes

```python
students = pd.DataFrame({'id_number' : [204, 202, 264, 136, 238],
                               'Name' : ['James Smith',
                                        'Maria Rodriguez',
                                        'David Smith', 
                                        'Patricia Garcia',
                                        'Jessica Hernandez'],
                               'Age' : [26, 22 ,24, 21, 25],
                            'Gender' : ['M', 'F', 'M', 'F', 'F'],
                    
                        })
```

```python
students
```

```python
professors = pd.DataFrame({'Prof_name' : ['Thomas Smith',
                                          'Margaret Garcia',
                                          'James Johnson', 
                                          'Richard Miller',
                                          'Margaret MOORE'],
                      'id_number' : [204, 264, 108, 238, 136],
                     'Department' : ['ECE', 'EE' ,'CSE', 'ME', 'CE'],
                      'Gender' : ['M', 'F', 'M', 'F', 'F'],
                         })
```

```python
professors1 = pd.DataFrame({'Prof_name' : ['Thomas Smith',
                                          'Margaret Garcia',
                                          'James Johnson', 
                                          'Richard Miller',
                                          'Margaret MOORE'],
                      'id_number' : [204, 264, 108, 238, 136],
                     'Department' : ['ECE', 'EE' ,'CSE', 'ME', 'CE'],
                      'Gender' : ['M', 'F', 'M', 'F', 'F'],
                         })
```

```python
pd.concat([students, professors])
```

```python
pd.concat([students, professors], ignore_index = True, sort = True)
```

```python
pd.merge(students, professors, how = 'outer')
```

```python
df = pd.merge(students, professors, how = 'outer', on = ['id_number', 'Gender'])
```

```python
pd.merge(students, professors, how = 'inner')
```

```python
pd.merge(students, professors, how = 'left')
```

```python
pd.merge(students, professors, how = 'right')
```

# Iterating over Data Frame

```python
space_org = {'Organizations_name' : ["CNES", "ISRO", "NASA", "JAXA", "CNSA"],
                   'Headquarters' : ["Paris", "Bangalore", "Washington",
                                     "Tokyo", "Beijing"],
                   'Country_name' : ["France", "India", "USA", "Japan", "China"],
                 'Formation_year' : [1961, 1969, 1958, 2003, 1993]
             }
```

```python
space_org_df = pd.DataFrame(space_org)
```

```python
space_org_df
```

```python
for i in space_org_df:
    print(i)
```

```python
for i in space_org_df.iterrows():
    print('row_index -> ', i[0])
    print(f'values at index {i[0]}-> \n', i[1])
    print()
```

```python
for i in space_org_df.iteritems():
    print('Column name = ', i[0])
    print(f'Value at coulmn {i[0]}', i[1])
    print()
```
